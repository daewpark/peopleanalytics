<h1> Bivariate (이변량) Analysis

<h3> Numeric - Numeric Analysis</h3>
<p>연속형 변수와 연속형 변수간의 관계를 파악하는 기술에 대해서 확인함

```python
#라이브러리 호출
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#기본 데이터 로드
df = pd.read_excel("resources/mean median mode.xlsx")
```

```python
df.plot(kind="scatter", x="age",y="salary_2018")
print("correlation between age & salary : ", df["salary_2018"].corr(df["age"]))
```

```python
#scatter plot matrix 만들기
df1 = df[["age","salary_2018","totalW_working_years","percent_hike"]]
df1.columns
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(df1,figsize = (20,10))
```

```python
#scatter plot 의 매트릭스를 만드는 방법은 seaborn 으로도 가능
import seaborn as sns
sns.pairplot(df1)
```

```python
sns.heatmap(df1.corr(), annot=True)
```

```python
#공분산 구하기
df2 = df.copy()
df2.cov() # 전체 변수간 공분산
```

```python
# 2개변수간 공분산
df2["age"].cov(df["salary_2018"])
```

```python
#Correlation
print("correlation between age & salary : ", df["salary_2018"].corr(df["age"]))
```

```python
#correlation matrix
df1 = df[["age","salary_2018","totalW_working_years","percent_hike"]]
df1.corr()
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df.corr(), annot=True)
```

<h3> Categorical - Categorical Analysis </h3>
<p> 카테고리- 카테고리 데이터간 관계를 확인하는 분석기법. 주로 Count 를 통해 파악한다.

```python
#Pandas cross tab
#엑셀에서 피벗테이블과 비슷한 역할을 함
#Self 를 크로스탭 앞에 쓰지 않는 것을 주의
df.info()
pd.crosstab(df["department"], df["gender"])
```

```python
#cross tab 비율로 구하기
#normalize = True 는 전체 중에서 해당 부분의 비율을 나타낸다. Sum(all) = 1
pd.crosstab(df["department"], df["gender"],normalize=True)
```

```python
#크로스 탭 비율 구하기에서 행비율 구하기
pd.crosstab(df["department"], df["gender"],normalize="index")
```

```python
#크로스 탭 비율 구하기에서 열비율 구하기
pd.crosstab(df["department"], df["gender"],normalize="columns")
```

```python
#Stacked bar chart 만들기
ax = pd.crosstab(df["gender"], df["department"]).plot.bar(stacked=True, figsize = (20,10))
```

```python
#타일형식 (모자이크) plot 만들기
from statsmodels.graphics.mosaicplot import mosaic
mosaic(df,["department","gender"])
plt.show()
```
